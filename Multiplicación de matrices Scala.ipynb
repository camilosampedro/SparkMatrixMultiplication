```scala
import org.apache.spark.mllib.linalg.{Vector, Vectors, Matrices}
```

```scala
import org.apache.spark.mllib.linalg.distributed.RowMatrix
```

```scala
def transpose(m: Array[Array[Double]]): Array[Array[Double]] = {
    (for {
        c <- m(0).indices
    } yield m.map(_(c)) ).toArray
  }
```

```scala
val file = sc.textFile("hdfs://localhost:9000/user/hadoop/matriz_n8")
```

```scala
val rows = file.map{line => 
    val values = line.split(' ').map(x=>(x.replace(",","").replace("[","").replace("]",""))).map(_.toDouble)
    Vectors.sparse(values.length, values.zipWithIndex.map(e=>(e._2, e._1)).filter(_._2 != 0.0))
}
```

```scala
file.take(2)
```

```scala
rows.take(2)
```

```scala
val rmat = new RowMatrix(rows)
```

```scala
val file2 = sc.textFile("hdfs://localhost:9000/user/hadoop/matriz_n9")
```

```scala
val dm1 = file2.map { line =>
    val values = line.split(' ').map(x=>(x.replace(",","").replace("[","").replace("]",""))).map(_.toDouble)
    Vectors.dense(values)
}
```

```scala
val ma = dm1.map(_.toArray).take(dm1.count.toInt)
```

```scala
val localMat = Matrices.dense( dm1.count.toInt,
                                 dm1.take(1)(0).size,
                                 transpose(ma).flatten)
```

```scala
val resultado = rmat.multiply(localMat).rows
```
